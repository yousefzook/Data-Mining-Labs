```python
import numpy as np
import pandas as pd
from IPython.display import display
```

### Data Loading and Exploration

```python
train = pd.read_csv("segmentation.data")
test = pd.read_csv("segmentation.test")

# adding the 2 files together
data = pd.concat([train, test])
```

```python
print("train dimensions:", train.shape)
print("test dimensions:", test.shape)
print("merged data dimensions:", data.shape)
display(data.head())
```

Here, the class column has no header. Let's add a name to it and reset index to be a natural index -i.e. 0,1,2..

```python
data.index.name = "class"
data = data.reset_index()

# show the new data frame header
display(data.head())
```

Now, let's get some facts about our data

```python
display(data.describe())
```

```python
boxplots = data.boxplot(rot=90, figsize=(20,16))
```

**Observations:** 
* There are no empty or missed values
* the data attributes are very distinct in ranges
* *`REGION-CENTROID-COL`* have a normal distribution as shown in the box plot figure and no outliers
* *`REGION-PIXEL-COUNT`* has sd = 0, which mean that it is a fixed value = 9. So this is a redundant dimension which we expect to be removed.
* *`SHORT-LINE-DENSITY-2`* and `SHORT-LINE-DENSITY-2` has a very small sd and the values = 0 till the 3rd quartile, it is also expected to be not useful as it don't contribute in the data variance, however it seems that the non zero data considered as outliers.
* *`VEDGE-MEAN`* and *`HEDGE-MEAN`* are similar to each others and as shown in the figure and the statstics table, they have outliers which affects their sd and mean.	
* *`HEDGE-SD`* and *`VEDGE-SD`* are similar to each others also. As shown in the figure and the statstics table, they have the most extreme outliers that extremely affect their sd and mean.
* *`SATURATION-MEAN`* seems to be a good attribute that is nearly normal distributed as shown in the statsics table
* some attributes have -ve values, for example *`EXRED-MEAN`* and *`EXGREEN-MEAN`*

**Conclusion**
* The data needs to be normalized
* There exist outliers, need to be removed
* Some attributes are considered redundancy so should be removed.
