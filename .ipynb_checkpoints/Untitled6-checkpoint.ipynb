Loading the iris dataset

```python
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data
Y = iris.target
```

Applying cosine similarity on the data

```python
from sklearn.metrics.pairwise import cosine_similarity

dist_matrix = cosine_similarity(X)
```

Plot the distance matrix , data and classes

```python
import matplotlib.pyplot as plt

# plt.plot(dist_matrix)

# print(1-dist_matrix)

plt.imshow(dist_matrix)
plt.show()

plt.imshow(1-dist_matrix)
plt.show()

plt.plot(X)
plt.show()

plt.plot(Y)
plt.show()
```

Get data for each class

```python
# Data is not shuffled , First 50 flowers belong to the same class , so the second and the third 50s

X1 = X[:50]
X2 = X[50:100]
X3 = X[100:150]
```

Plot these classes' features

```python
plt.plot(X1)
plt.show()

plt.plot(X2)
plt.show()

plt.plot(X3)
plt.show()
```

Plot these classes' features using histogram

```python
fig, axs = plt.subplots(1, 3, sharey=True, tight_layout=True)

n_bins = 10

axs[0].hist(X1, bins=n_bins)
axs[1].hist(X2, bins=n_bins)
axs[2].hist(X3, bins=n_bins)
```

Plot every 2 attributes together in scatter plot

```python
import numpy as np

splt_arr = np.hsplit(X, 4)

for i in range (len(splt_arr)):
  for j in range (i+1, len(splt_arr)):
#     arr_two_attrs = np.column_stack((splt_arr[i],splt_arr[j]))
#     plt.plot(arr_two_attrs)
    plt.scatter(splt_arr[i],splt_arr[j])
    plt.show()
```

Plot every 3 attributes together in scatter plot

```python
from mpl_toolkits.mplot3d import axes3d



for i in range (len(splt_arr)):
  for j in range (i+1, len(splt_arr)):
    for k in range (j+1, len(splt_arr)):
      fig = plt.figure()
      ax = fig.add_subplot(1, 1, 1, axisbg="1.0")
      ax = fig.gca(projection='3d')
      ax.scatter(splt_arr[i],splt_arr[j],splt_arr[k])
      plt.show()
```
